# Matplotilib의 다양한 플롯

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic') 
plt.rcParams['axes.unicode_minus'] = False
```

## 1. Bar Chart

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
np.random.seed(2021)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

## 2. 파이차트
    - 카테고리별 값의 상대적 비교시

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart", fontsize = 15)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

## 4. 산점도(scatter plot) - 굉장히 중요하다!!
    - 두 개의 실수 데이터 집합의 상관관계를 살펴볼때 사용

```python
np.random.seed(0)
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.title("Scatter Plot")
plt.scatter(X, Y)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```

### Anscombe's Quartet
    - 선형회귀선: y = 3.00 + 0.500 * x

```python
import seaborn as sns
ans = sns.load_dataset("anscombe")
ans.shape
```

```python
ans.head()
```

```python
ans.groupby("dataset").describe()
```

```python
ds1 = ans[ans.dataset == "I"]
ds1
```

#### dataset 1 에 대해서 산점도와 회기선

```python
plt.scatter(ds1.x, ds1.y)
plt.grid()
plt.title("dataset 1")
plt.show()
```

```python
plt.scatter(ds1.x, ds1.y)
plt.plot([2,16],[4,11], c="r", lw="2")
plt.grid()
plt.title("dataset 1")
plt.show()
```

#### 4개의 데이터셋 모두에 대해서 그림

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

for i, roman in enumerate(["I","II","III","IV"]):
    ds = ans[ans.dataset == roman]
    ax = axes[i//2, i%2]
    ax.scatter(ds.x, ds.y)
    ax.grid()

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6), sharex=True, sharey=True)

for i, roman in enumerate(['I','II','III','IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[i//2, i%2]
    ax.scatter(ds.x, ds.y)
    ax.plot([3,20],[4.5,13], "r-", lw=2)
    ax.set_title(f"Dataset {roman}")
    ax.grid()

fig.suptitle("Anscombe's Quartet", fontsize=16, y=1.02)
fig.tight_layout()
plt.show()
```

### 6. box plot

```python
iris = sns.load_dataset("iris")
iris.head()
```

### setosa품종의 각 feature의 박스 플롯

```python
setosa = iris[iris.species == "setosa"]
plt.figure(figsize=(8,10))
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
            setosa.petal_length, setosa.petal_width],
            labels = ["sepal_length","sepal_width","petal_length","petal_width"])
plt.title("setosa 품종의 각 feature별 박스플롯")
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize=(12,8), sharey= True)

for i, species in enumerate(iris.species.unique()):
    sub = iris[iris.species == species]
    ax = axes[i]
    ax.boxplot([setosa.sepal_length, setosa.sepal_width,setosa.petal_length, setosa.petal_width],
            labels = ["sepal_length","sepal_width","petal_length","petal_width"])
    ax.set_title(species)

plt.suptitle("setosa 품종의 각 feature별 박스플롯")
plt.tight_layout()
plt.show()
```
