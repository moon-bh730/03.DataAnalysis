# 라인 플롯

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic') 
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.title("라인 플롯")
plt.plot([1,4,9,16])        # y = (x + 1) ** 2
plt.grid()
plt.show()
```

```python
plt.title("라인 플롯")
xs = np.linspace(0, 3, 301)
ys = (xs + 1) ** 2
plt.plot(xs, ys)        # y = (x + 1) ** 2
plt.grid()
plt.show()
```

### 스타일 지정

```python
# 색(r), 마커=점(s), 선스타일(--)
plt.title("'rs --' 스타일의 plot")
plt.plot([10, 20, 30, 40],[1,4,9,16],"rs--")
plt.grid()
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Fixing random state for reproducibility
np.random.seed(19680801)

# Compute pie slices
N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
colors = plt.cm.viridis(radii / 10.)

ax = plt.subplot(projection='polar')
ax.bar(theta, radii, width=width, bottom=0.0, color=colors, alpha=0.5)

plt.show()
```

### 그림 범위 지정

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)
plt.ylim(-10, 30)

plt.show()
```

### 틱 설정

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid()
plt.show()
```

### 여러개의 선 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

```python
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--')
plt.plot(t, 0.5 * t**2, 'bs:')
plt.plot(t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

### 범계(Legend)

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
#plt.plot(X, C, ls="--", label="cosine")
#plt.plot(X, S, ls=":", label="sine")
plt.plot(X, C, label="cosine")      #라벨을 지정해야 범례에 표시됨
plt.plot(X, S, label="sine")        #라벨을 지정해야 범례에 표시됨
#plt.legend(loc=2)
#plt.legend(loc="lower center")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

### x축, y축 라벨, 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.grid()
plt.show()
```
