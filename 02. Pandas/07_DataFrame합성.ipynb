# 데이터프레임 합성

```python
import numpy as np
import pandas as pd
```

## mrege(sql의 Join)

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
# inner join(교집합)
pd.merge(df1, df2)      # how = "inner" (default)
```

```python
# left outer join
pd.merge(df1, df2, how="left")
```

```python
# right outer join
pd.merge(df1, df2, how="right")
```

```python
# full outer join
pd.merge(df1, df2, how="outer")
```

```python
df1 = pd.DataFrame({
    '고객명': ['춘향', '춘향', '몽룡'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터': ['20000', '30000', '100000']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명': ['춘향', '몽룡'],
    '데이터': ['여자', '남자']})
df2
```

```python
pd.merge(df1, df2, on="고객명")
```

```python
df1 = pd.DataFrame({
    '이름': ['영희', '철수', '철수'],
    '성적': [1, 2, 3]})
df1
```

```python
df2 = pd.DataFrame({
    '성명': ['영희', '영희', '철수'],
    '성적': [4,5,6]})
df2
```

```python
pd.merge(df1, df2, left_on='이름', right_on="성명")
```

## join method

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1.set_index("고객번호", inplace=True)         #inplace=True 해당 데이터 프레임을 정렬하라
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2.set_index("고객번호", inplace=True)
df2
```

```python
df1.join(df2)
```

## concat 함수 - 데이터 연결기능
    - 단순히 두 시리즈나 데이터프레임을 연결하기 때문에 인덱스 값이 중복될 수 있다.
    - union과 같은 개념인듯?

```python
s1 = pd.Series([0, 1], index=['A', 'B'])
s1
```

```python
s2 = pd.Series([2, 3, 4], index=['A', 'B', 'C'])
s2
```

```python
pd.concat([s1, s2])
```

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(3, 2),
    index=['a', 'b', 'c'],
    columns=['데이터1', '데이터2'])
df1
```

```python
df2 = pd.DataFrame(
    5 + np.arange(4).reshape(2, 2),
    index=['a', 'c'],
    columns=['데이터3', '데이터4'])
df2
```

```python
pd.concat([df1, df2],axis=1)
```

### 연습문제 2
    - 어느 회사의 전반기(1월 ~ 6월) 실적을 나타내는 데이터프레임과 후반기(7월 ~ 12월) 실적을 나타내는 데이터프레임을 만든 뒤 합친다. 
    - 실적 정보는 “매출”, “비용”, “이익” 으로 이루어진다. (이익 = 매출 - 비용).
    - 또한 1년간의 총 실적을 마지막 행으로 덧붙인다.

```python
df1 = pd.DataFrame(np.random.randint(100,1000, size=(6,2)),
                    columns=["매출","비용"],
                    index=["1월","2월","3월","4월", "5월","6월"])
df1["이익"] = df1.매출 - df1.비용
df1
```

```python
df2 = pd.DataFrame(np.random.randint(100,1000, size=(6,2)),
                    columns=["매출","비용"],
                    index=["7월","8월","9월","10월", "11월","12월"])
df2["이익"] = df2.매출 - df2.비용
df2
```

```python
df = pd.concat([df1, df2])
df
```

```python
# 합구하기
tdf = pd.DataFrame(df.sum()).T
tdf
```

```python
tdf.index = ["합계"]
tdf
```

```python
df = pd.concat([df, tdf])
df
```
