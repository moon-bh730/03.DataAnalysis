```python
import pandas as pd
import numpy as np

np.random.seed(88)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
# linspace 자동 배열생성 함수
# Linearly Spaced 줄임말
# (시작값, 끝값, 증감수)
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))], axis=1)
df.iloc[3, 3] = np.nan
df.iloc[0, 2] = np.nan

df.head()
```

```python
df.style.highlight_max(color='green').highlight_min(color='pink').highlight_null(null_color='red')
```

```python
df.style.background_gradient()
```

```python
# * 파라미터는 가변형 변수(몇개의 파라미터를 받을지, 안받을지 모를경우)
# *[para] - 튜플/리스트, **kwargs[para] - 딕셔너리
df.style.set_properties(**{'border': '1.3px solid green', 'color': 'magenta'})
```

```python
df.style.bar()
```

```python
df.style.bar(color="green")
```

```python
# 표시할 소숫점 자리수 지정
df.style.set_precision(2)
```

```python
# 제목 캡션 
df.style.set_caption('This is Dataframe styling demo').set_precision(2).background_gradient()
```

```python
df.style.hide_index()
```

```python
df.style.hide_columns('B')
```

```python
# 소수이하 3자리 - null에도 %가 붙는다
df.style.format("{:.3%}")   
```

```python
# null을 &&으로 표현하도록 지정
df.style.format("{:.3%}", na_rep = "&&")   
```

```python
styles = [
	dict(selector = "tr:hover",props=[("background","#f4f4f4")]),
	dict(selector = "th", props=[("color","#fff")
					,("border","1px solid #eee")
					,("padding","12px 35px")
					,("border-collapse","collapse")
					,("background","#00cccc")
					,("text-transform","uppercase")
					,("font-size","18px")]),
	dict(selector = "td", props=[("color","#999")
					,("border","1px solid #eee")
					,("padding","12px 35px")
					,("border-collapse","collapse")
					,("font-size","15px")]),
	dict(selector = "table", props=[("font-family","Arial")
					,("margin","25px auto")
					,("border-collapse","collapse")
					,("border","1px solid #eee")
					,("border-bottom","2px solid #00cccc")]),
	dict(selector = "caption", props=[("caption-side","bottom")])
]
```

```python
df.style.set_table_styles(styles).set_caption("Image by author (Made in Pandas)").highlight_max().highlight_null(null_color="red")
```

```python
# engine='openpyxl' - 적용한 스타일을 반영하여 저장하는 옵션
df.style.set_precision(2).background_gradient().hide_index().to_excel('styled.xlsx', engine='openpyxl')
```
