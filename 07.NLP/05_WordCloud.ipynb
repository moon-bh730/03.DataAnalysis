# 1. Alice in Wonderland

```python
maskfile = Image.open('data/Alice_mask.png')
```

```python
with open("data/alice.txt") as fp:
    text = fp.read()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words.add('said')
```

```python
wc = WordCloud(background_color='white',
            max_words=1000, stopwords=stop_words)
wc.generate(text)
```

```python
type(wc.words_)
```

```python
keys = list(wc.words_.keys())
values = list(wc.words_.values())
for i in range(10):
    print(keys[i], ':', values[i])
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
mask = np.array(Image.open('data/Alice_mask.png'))
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color='white', mask=mask,
               max_words=1000, stopwords=stop_words)
wc.generate(text)
```

```python
plt.figure(figsize=(10,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
with open("data/A_new_hope.txt") as fp:
    text = fp.read()
```

```python
maskfile = Image.open('data/Stormtrooper_mask.png')
```

```python
with open("data/A_new_hope.txt") as fp:
    text = fp.read()
text = text.replace('HAN', 'Han')
text = text.replace("LUKE'S", 'Luke')
```

```python
from wordcloud import STOPWORDS
stop_words = set(STOPWORDS)
stop_words.add('int')
stop_words.add('ext')
len(stop_words)
```

```python
mask = np.array(Image.open('data/Stormtrooper_mask.png'))
wc = WordCloud(mask=mask, max_words=1000, stopwords=stop_words,
               margin=10, random_state=2021)
wc.generate(text)
default_colors = wc.to_array
```

```python
import random
def grey_color(word, font_size, position,
               orientation, random_state=None, **kwargs):
    return f'hsl(0, 0%, {random.randint(60,100)}%)'
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=grey_color, random_state=2021), 
           interpolation='bilinear')
plt.axis('off')
plt.show()
```
