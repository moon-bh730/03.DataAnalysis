```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic') 
plt.rcParams['axes.unicode_minus'] = False
```

```python
filename = "유투브_순위.csv"
```

```python
df = pd.read_csv(filename)
df.head()
```

```python
# 구독자수 --> 구독자수2(정수)
# df["구독자수2"] = df.구독자수.apply(lambda x: int(x.replace(",","")))
# print(df.구독자수.apply(lambda x: x.replace(",","")))
df["구독자수2"] = df.구독자수
df.head()
```

```python
def str2int(x):
    return int(x.replace(",",""))
```

```python
df["비디오수2"] = df.비디오수.apply(str2int)
df.head()
```

```python
df.head()
```

```python
# df.sort_values(by="비디오수",ascending=False).head(10)
df_tmp01 = df.sort_values(by="비디오수",ascending=False).head(5)
df_tmp01.head()
```

```python
plt.bar(df_tmp01["채널명"],df_tmp01["비디오수"])
plt.show()
```

```python
df2 = df.sort_values(by='비디오수', ascending=False).head(20)

plt.barh(df2['채널명'], df2['비디오수'])
plt.show()
```

```python
df2 = df[['채널명', '비디오수']].sort_values(by='비디오수', ascending=False)
df2.set_index('채널명', inplace=True)
df2.head(20).sort_values(by='비디오수').plot(kind='barh', grid=True, figsize=(12,8))
plt.title('비디오수 Top20 채널')
plt.show()
```

```python
import seaborn as sns
df2 = df[['채널명', '비디오수']].sort_values(by='비디오수', ascending=False)
sns.barplot(y='채널명', x='비디오수', data=df2.head(20))
plt.title('비디오수 Top20 채널')
plt.show()
```

- 조회수 기준 top20 채널 시각화

```python
df2 = df[['채널명', '조회수']].sort_values(by='조회수', ascending=False)
sns.barplot(y='채널명', x='조회수', data=df2.head(20))
plt.title('조회수 Top20 채널')
plt.show()
```

- 카테고리 수 분포

```python
df.카테고리.value_counts()
```

```python
df[["카테고리","채널명"]].groupby("카테고리").count()
```

```python
df3 = df['카테고리'].value_counts().to_frame()

plt.figure(figsize=(12,8))
plt.pie('카테고리', labels=df3.index, data=df3, autopct='%.1f%%')
plt.title('카테고리별 채널 수')
plt.show()
```

```python
df4 = df[['카테고리','구독자수2']].groupby('카테고리').agg(['count', 'sum'])
df4.columns = ['채널수','구독자수_합계']
df4.sort_values(by='구독자수_합계', ascending=False, inplace=True)
df4
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='구독자수_합계', y=df4.index, data=df4)
plt.title('카테고리별 구독자수 합계')
plt.grid()
plt.show()
```

```python
df_new = df[['카테고리', '구독자수2']].groupby('카테고리').agg(['count','sum']) \
                .reset_index().sort_values(by=('구독자수2', 'sum'), ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x= ('구독자수2', 'sum'), y='카테고리', data=df_new)
plt.title('카테고리별 구독자수 합계', size=15)
plt.xlabel('구독자수(단위:억)', size=12)
plt.ylabel('카테고리', size=12)
plt.show()
```

```python
df4 = df.groupby('카테고리').sum().sort_values(by='구독자수2',ascending=False)

plt.figure(figsize=(12,8))
sns.barplot(x='구독자수2', y=df4.index, data=df4)
plt.xlabel('구독자 합계')
plt.title('카테고리별 구독자 수')
plt.show()
```
